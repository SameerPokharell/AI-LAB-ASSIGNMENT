```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/earthquake_1995-2023.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='alert')
df['magnitude'].value_counts()
```

```python
# pie chart

df['alert'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Numberical Value

# Histogram treat as categorical data distribution of data

plt.hist(df['depth'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['depth'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['depth'])
```
